**Sistema de Búsqueda Bibliográfica para Oftalmólogos**

**Introducción**

Nombre del proyecto

Sistema de Búsqueda Bibliográfica para Oftalmólogos: Resumen Automatizado de Literatura Científica

**Presentación del problema a abordar**

Los profesionales de la oftalmología enfrentan un desafío significativo: mantenerse actualizados con la creciente cantidad de literatura científica en su campo. La investigación oftalmológica avanza rápidamente, con miles de artículos publicados anualmente en revistas especializadas. Este volumen de información presenta varias problemáticas:

Sobrecarga de información: La cantidad de artículos científicos publicados en oftalmología ha crecido exponencialmente, haciendo imposible que los especialistas revisen toda la literatura relevante.

Barrera idiomática: Gran parte de la investigación de vanguardia se publica en inglés, lo que representa una dificultad adicional para oftalmólogos hispanohablantes.

Tiempo limitado: Los oftalmólogos dividen su tiempo entre atención clínica, cirugías, docencia e investigación, dejando poco espacio para revisar literatura científica extensa.

Complejidad técnica: Los abstracts científicos suelen ser densos y técnicos, requiriendo tiempo considerable para extraer los puntos clave relevantes para la práctica clínica.

La relevancia de desarrollar una solución a este problema es múltiple:

Mejora de la práctica clínica: El acceso eficiente a los últimos avances puede traducirse directamente en mejores resultados para los pacientes.

Optimización del tiempo: Una herramienta que condense información clave permitiría a los oftalmólogos mantenerse actualizados sin sacrificar tiempo de atención clínica.

Democratización del conocimiento: Reducir la barrera idiomática facilita que especialistas en diferentes regiones accedan a investigación de vanguardia.

Toma de decisiones basada en evidencia: Facilitar el acceso a literatura actualizada promueve la aplicación de la medicina basada en evidencia.
Desarrollo de la propuesta de solución

El Sistema de Búsqueda Bibliográfica para Oftalmólogos se vincula directamente con el desarrollo de modelos de IA, específicamente utilizando modelos de lenguaje grande (LLMs) para procesar, analizar y sintetizar literatura científica especializada.

Nuestra solución aprovecha el poder de los modelos de IA avanzados (Claude 3.7 Sonnet) para realizar una tarea compleja que combina comprensión de lenguaje especializado, extracción de información relevante y generación de resúmenes concisos manteniendo el rigor científico. El sistema utiliza técnicas de prompting específicas para optimizar la generación de resúmenes oftalmológicos:

**Técnicas de prompting implementadas:**

Few-Shot Learning: Proporcionamos al modelo ejemplos específicos de artículos oftalmológicos y sus correspondientes resúmenes para "enseñarle" el formato y estilo deseados. Esta técnica ha demostrado mejorar significativamente el rendimiento de los LLMs en tareas especializadas.

Instrucciones estructuradas: El prompt incluye directrices específicas sobre el formato, longitud y elementos críticos a incluir en el resumen (objetivo, métodos, resultados principales y conclusiones).

Contextualización por dominio: Incorporamos contexto específico de oftalmología para ayudar al modelo a entender la importancia relativa de diferentes elementos en un artículo científico de esta especialidad.

**Herramientas y tecnologías**

Técnicas de Prompting Principales

Few-Shot Learning: Esta técnica ha sido seleccionada como método principal porque ha demostrado ser especialmente efectiva para tareas especializadas como el resumen de literatura médica. Al proporcionar ejemplos concretos de abstracts oftalmológicos y sus resúmenes, el modelo aprende el formato específico, la terminología relevante y la estructura deseada. Justificación: Los artículos oftalmológicos contienen terminología muy especializada y una estructura particular. Few-shot learning permite "mostrar" al modelo exactamente cómo queremos que se realice el resumen, en lugar de solo describirlo.

Instrucciones Estructuradas: Los prompts incluyen instrucciones detalladas sobre qué elementos deben priorizarse (objetivo, métodos, resultados principales, conclusiones) y cómo debe estructurarse el resumen.

Justificación: La estructura clara asegura consistencia entre diferentes resúmenes y garantiza que se incluyan todos los componentes críticos para la comprensión del artículo.

Parámetros Optimizados: Utilizamos una temperatura baja (0.2) para favorecer respuestas más deterministas y precisas. Justificación: En el contexto médico, la precisión es crítica, y una temperatura baja reduce la creatividad innecesaria que podría introducir imprecisiones.

```python

# Instalar la biblioteca de Anthropic si no la tienes
!pip install anthropic

# Importar bibliotecas necesarias
import anthropic
import os
from google.colab import userdata

# Obtener la API key desde los secretos de Colab (primero debes configurarla)
try:
    api_key = userdata.get('ANTHROPIC_API_KEY')
    os.environ["ANTHROPIC_API_KEY"] = api_key

    # Configurar el cliente
    client = anthropic.Anthropic(api_key=os.environ["ANTHROPIC_API_KEY"])

    # Verificar la conexión
    response = client.messages.create(
        model="claude-3-sonnet-20240229",
        max_tokens=100,
        messages=[
            {"role": "user", "content": "Hola, ¿estás funcionando?"}
        ]
    )
    print("✅ API configurada correctamente!")
    print("Respuesta de Claude:", response.content[0].text)
except Exception as e:
    print(f"❌ Error: {e}")
    print("Para configurar tu API key como secreto en Colab:")
    print("1. Haz clic en el ícono 🔑 en el panel izquierdo")
    print("2. Agrega una nueva variable secreta:")
    print("   - Nombre: ANTHROPIC_API_KEY")
    print("   - Valor: tu-api-key-de-anthropic")
    print("3. Ejecuta nuevamente esta celda")

# Importar las bibliotecas necesarias
import time
from IPython.display import display, HTML
import pandas as pd
import re
from datetime import datetime

# Verificar que la API key está configurada correctamente
# (asumiendo que ya la configuraste con uno de los métodos anteriores)
client = anthropic.Anthropic(api_key=os.environ["ANTHROPIC_API_KEY"])

# Definir el modelo a utilizar (usando el más reciente disponible)
MODEL = "claude-3-sonnet-20240229"  # Actualizar según sea necesario

# Función para llamar a la API de Claude
def query_claude(prompt, temperature=0.2, max_tokens=1000):
    """
    Envía una consulta a Claude y devuelve la respuesta.

    Args:
        prompt (str): El texto que se enviará a Claude
        temperature (float): Controla la aleatoriedad (0.0-1.0)
        max_tokens (int): Número máximo de tokens en la respuesta

    Returns:
        str: La respuesta generada por Claude
    """
    try:
        response = client.messages.create(
            model=MODEL,
            max_tokens=max_tokens,
            temperature=temperature,
            messages=[
                {"role": "user", "content": prompt}
            ]
        )
        return response.content[0].text
    except Exception as e:
        print(f"Error al llamar a Claude: {e}")
        return f"[Error: No se pudo generar el resumen. Detalles: {e}]"

# Función de Few-Shot Prompting para resúmenes
def summarize_article_few_shot(title, abstract, authors, journal, publish_date):
    """
    Genera un resumen conciso de un artículo científico oftalmológico
    utilizando Few-Shot Prompting.

    Args:
        title (str): Título del artículo
        abstract (str): Texto completo del abstract
        authors (str): Autores del artículo
        journal (str): Nombre de la revista
        publish_date (str): Fecha de publicación

    Returns:
        str: Resumen generado por Claude
    """
    prompt = f"""
    Crea un resumen conciso en español del siguiente artículo científico para oftalmólogos.

    Título: {title}
    Autores: {authors}
    Revista: {journal}
    Fecha: {publish_date}
    Abstract: {abstract}

    Aquí hay algunos ejemplos de resúmenes concisos de artículos oftalmológicos:

    Ejemplo 1:
    Título: "Efficacy of intravitreal injections in ARMD"
    Abstract: "Long study on age-related macular degeneration treatment with anti-VEGF agents..."
    Resumen: "Este estudio evaluó la eficacia de inyecciones intravítreas de anti-VEGF en pacientes con DMAE durante 24 meses. Los resultados mostraron una mejora significativa en la agudeza visual en el 67% de los pacientes tratados, con efectos adversos mínimos. Los autores recomiendan un régimen de tratamiento individualizado basado en la respuesta inicial."

    Ejemplo 2:
    Título: "New diagnostic techniques for glaucoma detection"
    Abstract: "Comparison of OCT, visual fields and new imaging methods..."
    Resumen: "Esta investigación comparó la precisión diagnóstica de la OCT, campos visuales y nuevos métodos de imagen para la detección temprana del glaucoma. Los resultados indican que la combinación de OCT y campos visuales ofrece la mayor sensibilidad (93%) y especificidad (89%). Los nuevos métodos de imagen mostraron promesa pero requieren validación adicional en poblaciones diversas."

    Ejemplo 3:
    Título: "Corneal crosslinking outcomes in keratoconus patients"
    Abstract: "Long-term follow-up of CXL procedures in progressive keratoconus..."
    Resumen: "Este estudio de seguimiento a 5 años evaluó la eficacia del crosslinking corneal (CXL) en 120 pacientes con queratocono progresivo. Se observó estabilización de la enfermedad en el 94% de los casos, con mejoras adicionales en la visión con corrección en un 48%. Los autores concluyen que el CXL es un tratamiento efectivo y duradero para detener la progresión del queratocono."

    Genera un resumen similar para el artículo proporcionado en 3-5 frases que destaque objetivo, metodología, hallazgos principales y conclusiones. Mantén todas las cifras y datos estadísticos clave. Usa términos técnicos apropiados para oftalmólogos sin explicarlos.

    El resumen debe estar en español y tener un estilo conciso y objetivo, adecuado para profesionales médicos.
    """

    print("Generando resumen del artículo...")
    return query_claude(prompt, temperature=0.3)

# Función para procesar múltiples artículos
def process_articles_batch(articles_list):
    """
    Procesa una lista de artículos y genera resúmenes para cada uno.

    Args:
        articles_list (list): Lista de diccionarios con datos de artículos

    Returns:
        pandas.DataFrame: DataFrame con los artículos y sus resúmenes
    """
    results = []

    for i, article in enumerate(articles_list):
        print(f"Procesando artículo {i+1}/{len(articles_list)}: {article['title']}")

        # Generar el resumen
        summary = summarize_article_few_shot(
            article["title"],
            article["abstract"],
            article["authors"],
            article["journal"],
            article["publish_date"]
        )

        # Calcular métricas
        original_length = len(article['abstract'])
        summary_length = len(summary)
        compression_ratio = round((1 - summary_length/original_length) * 100, 1)

        # Guardar resultados
        results.append({
            "title": article['title'],
            "authors": article['authors'],
            "journal": article['journal'],
            "publish_date": article['publish_date'],
            "abstract": article['abstract'],
            "abstract_length": original_length,
            "summary": summary,
            "summary_length": summary_length,
            "compression_ratio": compression_ratio
        })

        # Pausa para no exceder límites de API
        if i < len(articles_list) - 1:
            print("Esperando 2 segundos antes del siguiente artículo...")
            time.sleep(2)

    # Convertir a DataFrame
    return pd.DataFrame(results)

# Función para mostrar un artículo y su resumen de forma atractiva
def display_article_summary(article_data):
    """
    Muestra el artículo y su resumen en un formato HTML atractivo.

    Args:
        article_data (dict): Diccionario con los datos del artículo y resumen
    """
    display(HTML(f"""
    <div style="border: 1px solid #ddd; padding: 20px; border-radius: 5px; margin-bottom: 20px;">
        <h2 style="color: #2c3e50;">{article_data['title']}</h2>
        <p><strong>Autores:</strong> {article_data['authors']}</p>
        <p><strong>Revista:</strong> {article_data['journal']} | <strong>Fecha:</strong> {article_data['publish_date']}</p>

        <h3 style="color: #2c3e50; margin-top: 20px;">Abstract Original</h3>
        <div style="background-color: #f8f9fa; padding: 15px; border-radius: 5px; margin-bottom: 15px;">
            {article_data['abstract']}
        </div>

        <h3 style="color: #2c3e50;">Resumen Generado</h3>
        <div style="background-color: #e8f4f8; padding: 15px; border-radius: 5px; border-left: 4px solid #3498db;">
            {article_data['summary']}
        </div>

        <div style="margin-top: 15px; font-size: 0.9em; color: #7f8c8d;">
            <p><strong>Métricas:</strong>
            Abstract original: {article_data['abstract_length']} caracteres |
            Resumen: {article_data['summary_length']} caracteres |
            Compresión: {article_data['compression_ratio']}%</p>
        </div>
    </div>
    """))

# Función para guardar resultados en CSV
def save_results_to_csv(df, filename=None):
    """
    Guarda los resultados en un archivo CSV.

    Args:
        df (pandas.DataFrame): DataFrame con los resultados
        filename (str, optional): Nombre del archivo. Si es None, se genera automáticamente.

    Returns:
        str: Ruta del archivo guardado
    """
    if filename is None:
        timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
        filename = f"resumenes_oftalmologia_{timestamp}.csv"

    df.to_csv(filename, index=False, encoding='utf-8-sig')
    print(f"Resultados guardados en: {filename}")
    return filename

# Ejemplo de uso con un artículo real
articulo_ejemplo = {
    "title": "Micropulse Transscleral Cyclophotocoagulation in Refractory Glaucoma",
    "authors": "Garcia R, Chen W, Lopez M, et al.",
    "journal": "Journal of Glaucoma Studies",
    "publish_date": "2024-02-10",
    "abstract": "Purpose: To evaluate the safety and efficacy of micropulse transscleral cyclophotocoagulation (MP-TSCPC) in patients with refractory glaucoma. Methods: This prospective interventional case series included 75 eyes of 75 patients with refractory glaucoma treated with MP-TSCPC using a standardized protocol. Intraocular pressure (IOP), medication burden, visual acuity, and complications were assessed at baseline and at 1 day, 1 week, 1 month, 3 months, 6 months, and 12 months postoperatively. Treatment success was defined as IOP reduction ≥20% from baseline with no additional glaucoma surgery. Results: Mean IOP decreased from 29.4±5.6 mmHg at baseline to 16.3±4.2 mmHg at 12 months (p<0.001), representing a 44.6% reduction. Treatment success was achieved in 71% of eyes at 12 months. Mean number of medications decreased from 4.1±0.8 to 2.8±1.1 at 12 months (p<0.001). No significant decrease in visual acuity was observed. Complications included transient inflammation (12%), hypotony (5.3%), and persistent ocular pain (2.7%). Conclusion: MP-TSCPC appears to be an effective and relatively safe procedure for patients with refractory glaucoma, achieving significant IOP reduction and decreased medication burden with minimal complications."
}

# Generar el resumen para el ejemplo
resumen_generado = summarize_article_few_shot(
    articulo_ejemplo["title"],
    articulo_ejemplo["abstract"],
    articulo_ejemplo["authors"],
    articulo_ejemplo["journal"],
    articulo_ejemplo["publish_date"]
)

# Crear un diccionario con todos los datos para mostrar
articulo_completo = {
    **articulo_ejemplo,
    "abstract_length": len(articulo_ejemplo["abstract"]),
    "summary": resumen_generado,
    "summary_length": len(resumen_generado),
    "compression_ratio": round((1 - len(resumen_generado)/len(articulo_ejemplo["abstract"])) * 100, 1)
}

# Mostrar el resultado de forma atractiva
display_article_summary(articulo_completo)

# También mostrar las métricas en texto plano
print("MÉTRICAS:")
print(f"Longitud del abstract original: {articulo_completo['abstract_length']} caracteres")
print(f"Longitud del resumen generado: {articulo_completo['summary_length']} caracteres")
print(f"Tasa de compresión: {articulo_completo['compression_ratio']}%")

# Demostración de procesamiento por lotes
# (Descomentar para usar)
"""
# Lista de artículos de ejemplo
articulos_ejemplo = [
    articulo_ejemplo,
    {
        "title": "Long-term Outcomes of Premium IOLs in Cataract Surgery",
        "authors": "Johnson K, Smith B, Williams T, et al.",
        "journal": "Journal of Cataract & Refractive Surgery",
        "publish_date": "2023-11-15",
        "abstract": "Purpose: To compare long-term visual outcomes and patient satisfaction with various premium intraocular lenses (IOLs) in cataract surgery. Methods: This retrospective cohort study evaluated 450 eyes of 225 patients who underwent bilateral cataract surgery with implantation of multifocal, extended depth of focus (EDOF), or trifocal IOLs between 2015 and 2020. Patients were followed for a minimum of 3 years. Outcome measures included uncorrected and corrected distance, intermediate, and near visual acuity; contrast sensitivity; presence of dysphotopsias; spectacle independence; and overall satisfaction. Results: At 3-year follow-up, all premium IOL groups demonstrated excellent distance visual acuity (mean 20/25 or better). The trifocal IOL group showed superior near vision (mean logMAR 0.05±0.10) compared to multifocal (0.15±0.12) and EDOF (0.25±0.15) groups (p<0.001). Intermediate vision was best with EDOF lenses. Spectacle independence was highest in the trifocal group (92%), compared to multifocal (83%) and EDOF (76%) groups. Dysphotopsias were reported by 24% of multifocal, 18% of trifocal, and 12% of EDOF patients, with decreasing severity over time. Overall satisfaction was high across all groups (>85%). Conclusion: Modern premium IOLs provide excellent visual outcomes at various distances with high rates of spectacle independence. Lens selection should be tailored to individual visual needs, with trifocal lenses favoring near vision and EDOF lenses offering better intermediate vision with fewer dysphotopsias."
    }
]

# Procesar los artículos
resultados_df = process_articles_batch(articulos_ejemplo)

# Mostrar cada artículo con su resumen
for i, row in resultados_df.iterrows():
    print(f"
--- Artículo {i+1}/{len(resultados_df)} ---")
    display_article_summary(row)

# Guardar los resultados
save_results_to_csv(resultados_df)
"""

# Información de uso para guía del usuario
print("--- GUÍA DE USO ---")
print("Para utilizar este sistema con tus propios artículos:")
print("1. Crea una lista de diccionarios con tus artículos siguiendo el formato del ejemplo")
print("2. Llama a process_articles_batch(tu_lista_de_articulos)")
print("3. Utiliza display_article_summary() para ver cada artículo con su resumen")
print("4. Guarda los resultados con save_results_to_csv()")

# Función más simple para ingresar artículos en Colab
# Función mejorada para ingresar artículos en Colab con resultados legibles
def ingresar_articulo_simple():
    from IPython.display import display, HTML, Markdown, clear_output
    import ipywidgets as widgets

    # Crear widgets para cada campo
    title_widget = widgets.Text(description="Título:", style={'description_width': '120px'}, layout=widgets.Layout(width='90%'))
    authors_widget = widgets.Text(description="Autores:", style={'description_width': '120px'}, layout=widgets.Layout(width='90%'))
    journal_widget = widgets.Text(description="Revista:", style={'description_width': '120px'}, layout=widgets.Layout(width='90%'))
    date_widget = widgets.Text(description="Fecha:", placeholder="YYYY-MM-DD", style={'description_width': '120px'}, layout=widgets.Layout(width='90%'))
    abstract_widget = widgets.Textarea(description="Abstract:", style={'description_width': '120px'}, layout=widgets.Layout(width='90%', height='200px'))

    output = widgets.Output()

    # Mostrar los widgets
    display(HTML("<h3>Ingresar Nuevo Artículo</h3>"))
    display(title_widget)
    display(authors_widget)
    display(journal_widget)
    display(date_widget)
    display(abstract_widget)

    # Función para mostrar el resumen de manera legible
    def mostrar_resumen_legible(articulo_data):
        """
        Muestra el artículo y su resumen en formato de texto plano para mejor legibilidad.
        """
        # Crear un texto formateado con Markdown para mejor legibilidad
        markdown_text = f"""
        ## {articulo_data['title']}

        **Autores:** {articulo_data['authors']}
        **Revista:** {articulo_data['journal']}
        **Fecha:** {articulo_data['publish_date']}

        ### Abstract Original

        {articulo_data['abstract']}

        ### Resumen en Español

        {articulo_data['summary']}

        ---

        **Métricas:**
        Abstract original: {articulo_data['abstract_length']} caracteres
        Resumen: {articulo_data['summary_length']} caracteres
        Compresión: {articulo_data['compression_ratio']}%
        """

        # Mostrar usando Markdown para mejor formato
        display(Markdown(markdown_text))

        # También imprimir el resumen en texto plano por si hay problemas con Markdown
        print("--- RESUMEN EN ESPAÑOL (TEXTO PLANO) ---")
        print(articulo_data['summary'])
        print("--------------------------------------------")

    # Función para el botón de enviar
    def on_submit_button_clicked(b):
        # Limpiar la salida anterior
        output.clear_output()

        # Recopilar datos del artículo
        articulo = {
            "title": title_widget.value,
            "authors": authors_widget.value,
            "journal": journal_widget.value,
            "publish_date": date_widget.value,
            "abstract": abstract_widget.value
        }

        # Validar datos
        if not all([articulo["title"], articulo["abstract"]]):
            with output:
                print("Error: El título y el abstract son obligatorios.")
            return

        # Generar el resumen
        with output:
            print("Generando resumen del artículo...")

        try:
            resumen = summarize_article_few_shot(
                articulo["title"],
                articulo["abstract"],
                articulo["authors"],
                articulo["journal"],
                articulo["publish_date"]
            )

            # Añadir información adicional
            articulo_completo = {
                **articulo,
                "abstract_length": len(articulo["abstract"]),
                "summary": resumen,
                "summary_length": len(resumen),
                "compression_ratio": round((1 - len(resumen)/len(articulo["abstract"])) * 100, 1)
            }

            # Mostrar el resultado de forma legible
            with output:
                mostrar_resumen_legible(articulo_completo)

        except Exception as e:
            with output:
                print(f"Error al generar el resumen: {str(e)}")

    # Botón para enviar
    submit_button = widgets.Button(description="Procesar Artículo", button_style='success')
    submit_button.on_click(on_submit_button_clicked)
    display(submit_button)

    # Área para mostrar resultados
    display(output)

# Ejecuta esta función directamente
ingresar_articulo_simple()



# Función para cargar artículos desde un archivo CSV
def load_articles_from_csv(filepath):
    """
    Carga artículos desde un archivo CSV.

    Args:
        filepath (str): Ruta al archivo CSV

    Returns:
        list: Lista de diccionarios con los datos de los artículos
    """
    try:
        df = pd.read_csv(filepath)
        required_columns = ["title", "authors", "journal", "publish_date", "abstract"]

        # Verificar que el CSV tenga las columnas necesarias
        for col in required_columns:
            if col not in df.columns:
                print(f"Error: El archivo CSV no contiene la columna '{col}'")
                return []

        # Convertir DataFrame a lista de diccionarios
        articles = df[required_columns].to_dict('records')
        print(f"Se cargaron {len(articles)} artículos desde el archivo CSV.")
        return articles

    except Exception as e:
        print(f"Error al cargar el archivo CSV: {e}")
        return []

# Función para crear un formulario interactivo para ingresar artículos (para Colab)
def create_article_form():
    """
    Crea un formulario interactivo en Colab para ingresar datos de artículos.

    Returns:
        dict: Diccionario con los datos del artículo ingresados en el formulario
    """
    from google.colab import output
    from IPython.display import display, HTML
    import ipywidgets as widgets

    # Crear widgets para el formulario
    title_widget = widgets.Text(description="Título:", width='100%')
    authors_widget = widgets.Text(description="Autores:", width='100%')
    journal_widget = widgets.Text(description="Revista:", width='100%')
    date_widget = widgets.Text(description="Fecha:", placeholder="YYYY-MM-DD", width='100%')
    abstract_widget = widgets.Textarea(description="Abstract:", rows=10, width='100%')

    # Crear el formulario
    form_items = [title_widget, authors_widget, journal_widget, date_widget, abstract_widget]
    form = widgets.VBox(form_items)

    # Variable para almacenar el resultado
    result = {}

    # Función para manejar el botón de enviar
    def on_submit_button_clicked(b):
        result["title"] = title_widget.value
        result["authors"] = authors_widget.value
        result["journal"] = journal_widget.value
        result["publish_date"] = date_widget.value
        result["abstract"] = abstract_widget.value

        # Mostrar vista previa de los datos
        display(HTML("<h3>Datos ingresados:</h3>"))
        for key, value in result.items():
            display(HTML(f"<p><strong>{key}:</strong> {value}</p>"))

        # Ocultar el formulario
        form.layout.display = 'none'

    # Crear y mostrar el botón de enviar
    submit_button = widgets.Button(description="Ingresar Artículo")
    submit_button.on_click(on_submit_button_clicked)

    # Mostrar el formulario y el botón
    display(form)
    display(submit_button)

    # Esperar a que el usuario complete el formulario
    while not result:
        time.sleep(0.1)

    return result

# Menú interactivo para usar en Colab
def interactive_menu():
    """
    Muestra un menú interactivo para usar el sistema en Colab.
    """
    from IPython.display import display, HTML, clear_output
    import ipywidgets as widgets

    articles_list = []
    results_df = None

    # Función para manejar la acción según la opción seleccionada
    def handle_option(option):
        nonlocal articles_list, results_df
        clear_output()

        if option == "1":  # Ingresar artículo manualmente
            display(HTML("<h2>Ingresar Artículo Manualmente</h2>"))
            article = input_new_article()
            articles_list.append(article)
            print(f"Artículo añadido. Total: {len(articles_list)} artículos.")

        elif option == "2":  # Usar formulario interactivo
            display(HTML("<h2>Formulario de Artículo</h2>"))
            article = create_article_form()
            articles_list.append(article)
            print(f"Artículo añadido. Total: {len(articles_list)} artículos.")

        elif option == "3":  # Cargar desde CSV
            display(HTML("<h2>Cargar Artículos desde CSV</h2>"))
            filepath = input("Ruta del archivo CSV: ")
            loaded_articles = load_articles_from_csv(filepath)
            if loaded_articles:
                articles_list.extend(loaded_articles)
                print(f"Total de artículos: {len(articles_list)}")

        elif option == "4":  # Procesar artículos
            if not articles_list:
                print("No hay artículos para procesar. Ingrese al menos un artículo.")
            else:
                display(HTML("<h2>Procesando Artículos</h2>"))
                print(f"Procesando {len(articles_list)} artículos...")
                results_df = process_articles_batch(articles_list)

                # Mostrar cada artículo procesado
                for i, row in results_df.iterrows():
                    display_article_summary(row)

        elif option == "5":  # Guardar resultados
            if results_df is None or results_df.empty:
                print("No hay resultados para guardar. Procese artículos primero.")
            else:
                display(HTML("<h2>Guardar Resultados</h2>"))
                filename = input("Nombre del archivo CSV (dejar en blanco para nombre automático): ")
                if not filename:
                    filename = None
                save_results_to_csv(results_df, filename)

        elif option == "6":  # Salir
            display(HTML("<h2>Saliendo del sistema</h2>"))
            print("¡Gracias por usar el sistema de resumen de artículos oftalmológicos!")
            return

        # Mostrar el menú nuevamente después de cada operación
        show_menu()

    # Función para mostrar el menú
    def show_menu():
        display(HTML("""
        <h2>Sistema de Resumen de Artículos Oftalmológicos</h2>
        <p>Seleccione una opción:</p>
        """))

        options = [
            "1. Ingresar artículo manualmente",
            "2. Usar formulario interactivo para ingresar artículo",
            "3. Cargar artículos desde CSV",
            "4. Procesar artículos y generar resúmenes",
            "5. Guardar resultados en CSV",
            "6. Salir"
        ]

        # Crear botones para cada opción
        for option_text in options:
            option_value = option_text[0]  # Tomar el número
            button = widgets.Button(description=option_text, layout=widgets.Layout(width='100%'))
            button.option = option_value
            button.on_click(lambda b: handle_option(b.option))
            display(button)

    # Mostrar el menú inicial
    show_menu()

# Ejemplo de artículos adicionales en español
articulos_ejemplo_es = [
    {
        "title": "Efectividad de la Terapia Anti-VEGF en Edema Macular Diabético",
        "authors": "Rodríguez A, Martínez B, González C, et al.",
        "journal": "Revista Española de Oftalmología",
        "publish_date": "2024-01-15",
        "abstract": "Objetivo: Evaluar la eficacia y seguridad a largo plazo de diferentes agentes anti-VEGF en el tratamiento del edema macular diabético (EMD). Métodos: Estudio retrospectivo multicéntrico que incluyó 315 ojos de 210 pacientes con EMD tratados con ranibizumab, aflibercept o bevacizumab durante un período de seguimiento de 36 meses. Se evaluaron los cambios en la agudeza visual mejor corregida (AVMC), el grosor macular central (GMC) medido por OCT, el número de inyecciones y los eventos adversos. Resultados: La AVMC mejoró de 58.2±12.3 a 70.5±11.2 letras ETDRS a los 36 meses en el grupo de aflibercept, de 56.8±13.1 a 68.2±12.4 letras en el grupo de ranibizumab, y de 57.5±11.8 a 66.7±13.1 letras en el grupo de bevacizumab (p<0.05 para todos los grupos vs. basal, p=0.041 entre grupos). El GMC se redujo significativamente en todos los grupos, siendo mayor la reducción con aflibercept (295±32 µm) que con ranibizumab (263±38 µm) o bevacizumab (241±45 µm) (p=0.023). El número medio de inyecciones fue menor con aflibercept (10.2±2.3) comparado con ranibizumab (11.5±2.8) y bevacizumab (13.2±3.1) durante los 36 meses (p<0.01). No se observaron diferencias significativas en los perfiles de seguridad entre los tres medicamentos. Conclusiones: Los tres agentes anti-VEGF fueron efectivos en el tratamiento del EMD, con mejoría sostenida de la AVMC y reducción del GMC. Aflibercept mostró ventajas en términos de eficacia anatómica y menor número de inyecciones necesarias durante el seguimiento a largo plazo."
    },
    {
        "title": "Resultados visuales tras implantación de lentes intraoculares trifocales en cirugía de catarata",
        "authors": "López J, García M, Fernández R, et al.",
        "journal": "Archivos de la Sociedad Española de Oftalmología",
        "publish_date": "2023-09-22",
        "abstract": "Objetivo: Evaluar los resultados visuales y la satisfacción de los pacientes tras la implantación bilateral de lentes intraoculares (LIO) trifocales en cirugía de catarata. Método: Se incluyeron 120 ojos de 60 pacientes sometidos a cirugía de catarata bilateral con implantación de LIO trifocal PanOptix® (Alcon). Se evaluaron la agudeza visual sin corrección (AVSC) y con corrección (AVCC) para visión lejana, intermedia y cercana, la sensibilidad al contraste, los fenómenos disfotópsicos y la satisfacción del paciente mediante un cuestionario validado a los 6 meses postoperatorios. Resultados: La AVSC de lejos, intermedia y cerca fue de 0.01±0.10, 0.05±0.12 y 0.12±0.15 logMAR, respectivamente. El 92% de los pacientes alcanzaron una AVSC de lejos ≤0.1 logMAR, el 88% una AVSC intermedia ≤0.1 logMAR y el 84% una AVSC de cerca ≤0.1 logMAR. El 94% de los pacientes reportaron independencia de gafas para todas las distancias. La sensibilidad al contraste fue satisfactoria, aunque mostró una ligera reducción en condiciones mesópicas. El 24% de los pacientes reportaron halos o deslumbramiento leve-moderado, que disminuyeron en severidad durante el periodo de seguimiento. La satisfacción global fue alta, con un 92% de pacientes calificando sus resultados como buenos o excelentes. Conclusiones: Las LIO trifocales PanOptix® proporcionan una excelente agudeza visual a todas las distancias con alta independencia de gafas. Aunque los fenómenos disfotópsicos son comunes inicialmente, su impacto en la satisfacción general es limitado y tienden a reducirse con el tiempo."
    }
]

# Código para crear un botón que inicie el menú interactivo (para Colab)
def show_start_button():
    """
    Muestra un botón para iniciar el sistema interactivo.
    """
    from IPython.display import display
    import ipywidgets as widgets

    start_button = widgets.Button(
        description="Iniciar Sistema Interactivo",
        button_style='success',
        icon='play'
    )

    def on_button_clicked(b):
        interactive_menu()

    start_button.on_click(on_button_clicked)
    display(start_button)

# Si estás ejecutando en Colab, descomentar la siguiente línea para mostrar el botón de inicio
show_start_button()
```
